# pydyneval tutorial

Most-common usage: compare two trajectories.

```python
import sys; sys.path.insert(0,'..')
import pydyneval as de, pydynwrap as dw, numpy as np
gold = dw.add_branch_trajectory(dw.wrap_data([f'c{i}' for i in range(30)]),
    np.tile(np.linspace(0,1,10),3), np.repeat(['A','B','C'],10))
pred = dw.add_branch_trajectory(dw.wrap_data([f'c{i}' for i in range(30)]),
    np.tile(np.linspace(0,1,10),3), np.repeat(['A','B','C'],10))
print(de.calculate_metrics(gold, pred))
```

## Individual metrics

```python
print('him  :', de.calculate_him(gold.milestone_network, pred.milestone_network))
print('edge :', de.calculate_edge_flip(gold.milestone_network, pred.milestone_network))
print('iso  :', de.calculate_isomorphic(gold.milestone_network, pred.milestone_network))
```
